```python
from sklearn import preprocessing
def prepare_data(df_data):
    select_cols = ["survived", "name", "pclass", "sex", "age", "sibsp", "parch", "fare", "embarked"]
    select_df_data = df_data[select_cols]

    select_df_data['age'] = select_df_data['age'].fillna(select_df_data['age'].mean()).astype(int)
    fare_mean = select_df_data["fare"].mean()
    select_df_data["fare"] = select_df_data["fare"].fillna(fare_mean)
    select_df_data["embarked"] = select_df_data["embarked"].fillna("S")
    select_df_data["embarked"] = select_df_data["embarked"].map({"C":0, "Q":1, "S":2, 0:0,1:1,2:2}).astype(int)
    select_df_data["sex"] = select_df_data["sex"].map({"female":0, "male":1, 0:0,1:1}).astype(int)
    select_df_data = select_df_data.drop("name", axis=1)
    
    ndarry_data = select_df_data.values
    features = ndarry_data[:, 1:]
    labels = ndarry_data[:,0]
  

    minmax_scale = preprocessing.MinMaxScaler(feature_range=(0,1))
    norm_features = minmax_scale.fit_transform(features)
    return norm_features, labels
```

```python
shuffled_select_data = df_data.sample(frac=1)
x_data,y_data = prepare_data(shuffled_select_data)
print(x_data.shape, y_data.shape)

train_size = int(len(x_data) * 0.8)
x_train = x_data[:train_size]
y_train = y_data[:train_size]
x_test = x_data[train_size:]
y_test = y_data[train_size:]
print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)
```

## 建立模型结构


```python
import tensorflow as tf
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(units = 64, 
                               input_shape=(7,),
                               use_bias=True,
                               kernel_initializer ='uniform',
                                bias_initializer = 'zeros',
                                activation='relu'
                               ))
model.add(tf.keras.layers.Dense(units= 32, activation='sigmoid'))
model.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
model.summary()
```

```python
model.compile(optimizer=tf.keras.optimizers.Adam(0.003),
             loss='binary_crossentropy',
             metrics=['accuracy'])
```

```python
logdir = './results/tb_results'
checkpoint_path = "./results/Tinanic.{epoch:02d}-{val_acc:.2f}.ckpt"
callbacks = [
    tf.keras.callbacks.TensorBoard(log_dir=logdir),
    tf.keras.callbacks.ModelCheckpoint(filepath= checkpoint_path,
                                      verbose=1,)
]
```

```python
import os
checkpoint_dir = os.path.dirname(checkpoint_path)
print(checkpoint_dir)
latest = tf.train.latest_checkpoint(checkpoint_dir)
print(latest)
```

```python
model = tf.keras.models.load_model('./results/Tinanic.100-0.85.h5')
```

```python
history = model.fit(x=x_train,
                    y=y_train,
                    validation_split=0.2,
                   epochs=100,
                   batch_size=40,
                    callbacks=callbacks,
                   verbose=1)
```

```python
import matplotlib.pyplot as plt

def visual_train_history(history, train_metric, validation_metric):
    plt.plot(history.history[train_metric])
    plt.plot(history.history[validation_metric])
    plt.ylabel(train_metric)
    plt.xlabel('epochs')
    plt.legend(['train','validation'], loc='lower right')
    plt.show()
visual_train_history(history, 'acc', 'val_acc')
visual_train_history(history, 'loss', 'val_loss')
```
